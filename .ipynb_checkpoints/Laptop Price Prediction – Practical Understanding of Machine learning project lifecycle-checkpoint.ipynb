```python
df_price_company=df[["Company", "Price"]]
```

```python
df_price_company.head()
```

```python
df_apple=df_price_company.loc[df_price_company['Company'] == 'Apple']
df_apple.Price.mean()
```

```python
means = df_price_company.groupby('Company')['Price'].mean()
type(means)
```

```python
means=pd.DataFrame(means)
means.head()
```

```python
means=means.sort_values('Price')
```

```python
ax = means.plot(kind='bar')
ax.set_xlabel('Company')
ax.set_ylabel('Price')
ax.set_title('Average Price')
plt.show()
```
